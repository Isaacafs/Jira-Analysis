## Limpeza dos dados

##### Abaixo, teremos uma sequência de códigos que tiveram como objetivo analisar o dataset e pegar as melhores informações possíveis com relação a problemas que ainda estão em andamento.

Carregamento e Visualização Inicial dos Dados

```python
import pandas as pd

# Caminho do arquivo (ajuste conforme o download)
csv_path = "./data/jira_issues.csv"

# Carregar CSV
df = pd.read_csv(csv_path)

# Visualizar colunas
print("Colunas disponíveis:", df.columns.tolist())
print(df.head())
```

Exibição do DataFrame Completo

```python
df
```

Contagem de Issues por Status

```python
# Quantidade de issues por status
print(df['status'].value_counts())
```

Exemplo de Limpeza e Filtragem por Status

```python
# Padronizar coluna 'status' (remover espaços, deixar primeira letra maiúscula)
df['status'] = df['status'].str.strip().str.title().fillna("Unknown")

# --- Escolher os status que deseja visualizar ---
status_escolhidos = ["Resolved"]  # você pode trocar por qualquer lista

# Criar DataFrame filtrado
df_filtrado = df[df['status'].isin(status_escolhidos)]

# Mostrar resultado
df_filtrado
```

Correção de Status com Base na Data de Resolução

```python
# Converter resolved para datetime (ignorar erros)
df['resolved'] = pd.to_datetime(df['resolved'], errors='coerce')

# Forçar status = Resolved se há data
df.loc[df['resolved'].notna(), 'status'] = "Closed"
```

Visualização do DataFrame Após Correção

```python
# --- Escolher os status que deseja visualizar ---
status_escolhidos = ["Closed"]  # você pode trocar por qualquer lista

# Criar DataFrame filtrado
df_filtrado = df[df['status'].isin(status_escolhidos)]

# Mostrar resultado
df_filtrado  # mostra só 20 linhas para observação
```

Verificação da Distribuição de Status Pós-Correção

```python
# Quantidade de issues por status
print(df['status'].value_counts())
```

Agrupamento de Status em Categorias Genéricas

```python
# Criar mapeamento
status_map = {
    # Aberto
    "To Do": "Open",
    "New": "Open",
    
    # Em andamento
    "In Progress": "In Progress",
    "Coding In Progress": "In Progress",
    "Analysis": "In Progress",
    "Investigating": "In Progress",
    "Qa In Progress": "In Progress",
    "Waiting For Feedback": "In Progress",
    "Continued": "In Progress",
    
    # Revisão/Teste
    "Submitted": "Review",
    "Patch Available": "Review",
    "Pull Request Sent": "Review",
    "Ready To Review": "Review",
    "Out For Review": "Review",
    "Patch Pending": "Review",
    
    # Em espera
    "On Hold": "Waiting",
    "Waiting For Infra": "Waiting",
    "Waiting For Upstream": "Waiting",
    "With Customer": "Waiting",
    "Awaiting Test Case": "Waiting",
    "Waiting For User": "Waiting",
    "Pending Upload To Patch Repository": "Waiting",
    "Requires Porting": "Waiting",
    
    # Outro
    "Unkown": "Other",
    "Testing": "Other",
    "Iteration": "Other"
}

# Normalizar status
df['status'] = df['status'].replace(status_map)

# Corrigir incoerências com a coluna resolved
df.loc[df['status'].isin(["Closed", "Resolved"]) & df['resolved'].isna(), 'status'] = "Open"

# Visualizar distribuição final
print(df['status'].value_counts())
```

Padronização da Coluna resolution

```python
# Padronizar coluna 'resolution' (remover espaços, deixar primeira letra maiúscula)
df['resolution'] = df['resolution'].str.strip().str.title().fillna("Unknown")

# --- Escolher os resolution que deseja visualizar ---
resolution_escolhidos = ["Migrated To Another Its"]  # você pode trocar por qualquer lista

# Criar DataFrame filtrado
df_filtrado = df[df['resolution'].isin(resolution_escolhidos)]

# Mostrar resultado
df_filtrado  # mostra só 20 linhas para observação
```

Verificação da Distribuição de resolution

```python
print(df['resolution'].value_counts())
```

Agrupamento de Tipos de Resolução

```python
resolution_map = {
    # Concluído
    "Fixed": "Done",
    "Complete": "Done",
    "Implemented": "Done",
    
    # Não corrigido / rejeitado / inválido
    "Deferred": "Later",
    "Not A Bug": "Not A Problem",
    
    # Duplicados
    "Duplicate Issue": "Duplicate",
    
    # Bugs não reproduzíveis
    "Cannot Reproduce Bug": "Cannot Reproduce",
}

df['resolution'] = df['resolution'].replace(resolution_map)

# Visualizar distribuição
print(df['resolution'].value_counts(dropna=False))
```

Análise da Coluna priority

```python
print(df['priority'].value_counts(dropna=False))
```

Tratamento de Valores Nulos em priority

```python
# Substituir NaN por "Optional"
df['priority'] = df['priority'].fillna("Optional")
```

Criação de um DataFrame Filtrado para Issues Abertas e Urgentes

```python
# Prioridades urgentes/importantes
urgent_priorities = ["Major", "Critical", "Blocker"]

# Colunas essenciais para análise
essential_cols = [
    "created",
    "status",
    "priority",
    "assignee_id",
    "reporter_id",
]

# Remover linhas onde assignee_id ou reporter_id são NaN
df_opened_issues = df.dropna(subset=["assignee_id", "reporter_id"])

# Criar DataFrame enxuto
df_opened_issues = df_opened_issues[essential_cols]

# Remover status "Closed" e filtrar prioridades urgentes ao mesmo tempo
df_opened_issues = df_opened_issues[(df_opened_issues['status'] != "Closed") & (df_opened_issues['priority'].isin(urgent_priorities))]
```

Exibição do Novo DataFrame

```python
df_opened_issues
```

Exportação do DataFrame de Issues Abertas para CSV

```python
# Exportar para CSV
df_opened_issues.to_csv("jira_issues_opened.csv", index=False)
print("CSV com status e resolution normalizados criado com sucesso!")
```
